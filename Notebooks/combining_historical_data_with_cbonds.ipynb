### Dividing the bond listings into their resspective quarters or weeks when gathered to try and approximate a date of creation for each bond.

```python
# create a main path so we just have to add the different path extensions to it
github_path = 'https://raw.githubusercontent.com/gabriellavinco/ECB-Bond-Purchases/main/Datasets/historical_lists_cspp/'
```

```python
# import additions to path
add_path = pd.read_csv('https://www.dropbox.com/s/zts8panmd1qbwx1/df_paths.csv?dl=1')
```

```python
# change df to list
add_path = add_path["colummn"].to_list()
```

```python
github_df = []
```

```python
# goes through all the path additions/extensions and grabs the csv from github then appends the df to a list
for path in add_path:
#     print(path)
    path_year = path[:4]
    path_quarter = path[5:7]
    week_gathered = path[-12:-4]
#     print(week_gathered)
    github_path = str(github_path)
    path = str(path)
    new_path = github_path + path
#     print(new_path)
    df = pd.read_csv(new_path, index_col=None, header=0, encoding='latin-1')
    df["historical_date"] = week_gathered
    df["year"] = path_year
    df["quarter"] = path_quarter
    github_df.append(df)
```

```python
# github_df
```

```python
# concatenating all the dfs in github_df list
historical_dfs = pd.concat(github_df)
```

```python
# check the df
historical_dfs.head()
```

```python
# import cbonds
cbonds = pd.read_csv('https://www.dropbox.com/s/vxt2z1f4dn1jz4o/HISTORICAL_BONDS.csv?dl=1')
```

### Historical Data divided into quarters, years, and the week for the dataset it was taken from

```python
# change name of df
all_historical_copies = historical_dfs
```

```python
# eliminate redundant columns and combine into one
all_historical_copies["ISIN_Code"] = all_historical_copies["ISIN_CODE"].astype(str) + all_historical_copies["ISIN"].astype(str)
all_historical_copies["Issuer_Name"] = all_historical_copies["ISSUER_NAME_"].astype(str) + all_historical_copies["ISSUER_NAME"].astype(str) + all_historical_copies["ISSUER"].astype(str)
all_historical_copies["Maturity_Date"] = all_historical_copies["MATURITY_DATE_"].astype(str) + all_historical_copies["MATURITY_DATE"].astype(str) + all_historical_copies["MATURITY DATE"].astype(str)
all_historical_copies["Coupon_Rate"] = all_historical_copies["COUPON_RATE_"].astype(str) + all_historical_copies["COUPON_RATE_*"].astype(str) + all_historical_copies["COUPON_RATE"].astype(str) + all_historical_copies["COUPON RATE"].astype(str)
```

```python
# deleting all the obsolete columns
all_historical_copies = all_historical_copies.drop(columns=['ISIN_CODE',
                                                            'ISSUER_NAME_',
                                                            'MATURITY_DATE_',
                                                            'COUPON_RATE_',
                                                            'COUPON_RATE_*',
                                                            'ISSUER_NAME',
                                                            'MATURITY_DATE',
                                                            'COUPON_RATE',
                                                            'Unnamed: 5',
                                                            'Unnamed: 6',
                                                            'Unnamed: 0',
                                                            'ISIN',
                                                            'ISSUER',
                                                            'MATURITY DATE',
                                                            'COUPON RATE'])
```

```python
# reset index
all_historical_copies = all_historical_copies.reset_index(drop = True)
```

```python
# all_historical_copies
```

```python
# remove any nan values that were addded to our info upon the combination from before
for row in range(len(all_historical_copies)):
    all_historical_copies.iloc[row, 4] = all_historical_copies.loc[row]["ISIN_Code"].replace('nan', '')
    all_historical_copies.iloc[row, 5] = all_historical_copies.loc[row]["Issuer_Name"].replace('nan', '')
    all_historical_copies.iloc[row, 6] = all_historical_copies.loc[row]["Maturity_Date"].replace('nan', '')
    all_historical_copies.iloc[row, 7] = all_historical_copies.loc[row]["Coupon_Rate"].replace('nan', '')
```

```python
# df check
all_historical_copies.head()
```

```python
# create lists to compare the isin codes to then create a df with more information
all_isin_codes = all_historical_copies["ISIN_Code"].tolist()
cbonds_isin_list = cbonds["ISIN"].tolist()
```

```python
included_isin = all_historical_copies.loc[all_historical_copies['ISIN_Code'].isin(cbonds_isin_list)]
```

```python
# sort values
included_isin = included_isin.sort_values(by=['ISIN_Code'])
```

```python
# get only unique values
unique = included_isin.ISIN_Code.unique()
```

```python
# create an empty df
value = pd.DataFrame()
```

```python
# if item is in the cbonds list then add
for item in unique:
    info = cbonds.loc[cbonds["ISIN"] == item]
    value = value.append(info, ignore_index = True)
```

```python
# rename column to match other df column name
value = value.rename(columns={"ISIN": "ISIN_Code"})
```

```python
# merge left and combine the info into one df
combo = included_isin.merge(value, how='left', on='ISIN_Code')
```

```python
# check df size
print(combo.shape)
```

```python
# sort by isin code then historical date to see for how long the bonds were present in the ECB data
combo = combo.sort_values(by=['ISIN_Code', "historical_date"])
```
